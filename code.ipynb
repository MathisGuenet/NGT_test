```python
import numpy as np
#I will store my new datas from all instruments in this dataFrame
Nav_per_Share_Matrix = pd.DataFrame(columns = ["1A","1B","1C","1D",'2A',"2B","2C","2D","2E","3A","3B"])
Nav_per_Share_Matrix
```

```python
data = pd.read_csv("ngt_software_engineer_test_example_data (clean columns).csv")
data.set_index( ['Subfund_Code', 'Share_Class_Code'],inplace = True)
data
```

```python
df1A = data.loc[1,"A"][:34]['NAV_Per_Share'].to_numpy()
df2A = data.loc[2,"A"][:34]['NAV_Per_Share'].to_numpy()
df3A = data.loc[3,"A"][:34]['NAV_Per_Share'].to_numpy()
df1B = data.loc[1,"B"][:34]['NAV_Per_Share'].to_numpy()
df2B = data.loc[2,"B"][:34]['NAV_Per_Share'].to_numpy()
df3B = data.loc[3,"B"][:34]['NAV_Per_Share'].to_numpy()
df1C = data.loc[1,"C"][:34]['NAV_Per_Share'].to_numpy()
df2C = data.loc[2,"C"][:34]['NAV_Per_Share'].to_numpy()
df1D = data.loc[1,"D"][:34]['NAV_Per_Share'].to_numpy()
df2D = data.loc[2,"D"][:34]['NAV_Per_Share'].to_numpy()
df2E = data.loc[2,"E"][:34]['NAV_Per_Share'].to_numpy()
df1A
```

```python
Nav_per_Share_Matrix['1A'] = df1A
Nav_per_Share_Matrix['2A'] = df2A
Nav_per_Share_Matrix['3A'] = df3A
Nav_per_Share_Matrix['1B'] = df1B
Nav_per_Share_Matrix['2B'] = df2B
Nav_per_Share_Matrix['3B'] = df3B
Nav_per_Share_Matrix['1C'] = df1C
Nav_per_Share_Matrix['2C'] = df2C
Nav_per_Share_Matrix['1D'] = df1D
Nav_per_Share_Matrix['2D'] = df2D
Nav_per_Share_Matrix['2E'] = df2E
Nav_per_Share_Matrix
```

```python
#Now that all our data are saved, we can use the function corr() from pandas library
#And finally we have our correlation matrix for NAV_Per_Share between those 11 instruments
Nav_per_Share_Matrix.corr()
```
